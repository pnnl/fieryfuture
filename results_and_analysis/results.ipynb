```python
df = pandas.DataFrame(index=['LR', 'RF', 'DNN', 'RNN'], columns=['$D_1$', '$D_2$','$D_3$','$D_4$'])
# dnn = pandas.read_csv('d4dnn.csv', header=None)
# rnn = pandas.read_csv('d4rnn.csv', header=None)
parts = range(1,5)
lr = [pandas.read_csv('d%slr.csv' % p, header=None) for p in parts] 
for k in lr:
    k.columns = ['lr', 'reduce' ,'mb', 'maxepochs', 'l2', 'randomseed','decayrate', 'decaysteps', 'prec', 'rec', 
              'fscore', 'auc', 'accuracy', 'testprec', 'testrec', 'testfscore', 
              'testauc', 'testaccuracy', 'score', 'stdscore']
rf = [pandas.read_csv('d%srandom_forest.csv' %p) for p in parts]
best = [rf[p-1].iloc[rf[p-1]['accuracy'].argmax()] for p in parts]
dnn = [pandas.read_csv('d%sdnn.csv' % p, header=None) for p in parts] 

for k in dnn:
    k.columns = ['lr' ,'mb', 'maxepochs', 'kp', 'randomseed', 'lay1', 'lay2', 'lay3', 'decayrate', 
               'decaysteps', 'prec', 'rec', 
              'fscore', 'auc', 'accuracy', 'testprec', 'testrec', 'testfscore', 
              'testauc', 'testaccuracy', 'score', 'stdscore']
rnn = [pandas.read_csv('d%srnn.csv' % p, header=None) for p in parts] 
for k in rnn:
    k.columns = ['lr' ,'mb', 'maxepochs', 'kp', 'randomseed', 'lay1', 'lay2', 'lay3', 'decayrate', 
               'decaysteps', 'prec', 'rec', 
              'fscore', 'auc', 'accuracy', 'testprec', 'testrec', 'testfscore', 
              'testauc', 'testaccuracy', 'score', 'stdscore']

for p in parts:
    df['$D_%s$' % p]['RF'] = best[p-1]['accuracy']
for p in [1,2,3,4]:
    df['$D_%s$' % p]['LR'] = (lr[p-1].iloc[lr[p-1]['accuracy'].argmax()])['accuracy']
for p in [1,2,3,4]:
    df['$D_%s$' % p]['DNN'] = (dnn[p-1].iloc[dnn[p-1]['accuracy'].argmax()])['accuracy']
for p in [1,2,3,4]:
    df['$D_%s$' % p]['RNN'] = (rnn[p-1].iloc[rnn[p-1]['accuracy'].argmax()])['accuracy']
```
